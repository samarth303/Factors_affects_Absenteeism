### Import Data

```python
import pandas as pd
```

```python
raw_csv_data = pd.read_csv('Absenteeism_data.csv')
```

```python
df = raw_csv_data.copy()
```

```python
df.head()
```

## Drop ID

```python
df = df.drop(['ID'],axis=1)
```

```python
df['Reason for Absence'].unique()
```

```python
len(df['Reason for Absence'].unique())
```

```python
reason_columns = pd.get_dummies(df['Reason for Absence'],drop_first=True)
```

```python
df = df.drop(['Reason for Absence'],axis=1)
```

### Group Similar Reason to reduce nummber of variables
- Reason_Type 1 corresponds to reasons related various diseases
- Reason_Type 2 corresponds to pregnancy and giving birth
- Reason_Type 3 corresponds to Poisoning
- Reason_Type 4 corresponds to ligh diseases

```python
reason_columns = pd.DataFrame(reason_columns)
reason_type1 = reason_columns.loc[:,1:14].max(axis=1)
reason_type2 = reason_columns.loc[:,15:17].max(axis=1)
reason_type3 = reason_columns.loc[:,18:21].max(axis=1)
reason_type4 = reason_columns.loc[:,22:].max(axis=1)
```

```python
df = pd.concat([df,reason_type1,reason_type2,reason_type3,reason_type4],axis=1)
df
```

```python
df.columns.values
```

```python
column_names = ['Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours','Reason_1', 'Reason_2', 'Reason_3', 'Reason_4']
```

```python
df.columns = column_names
df.head()
```

```python
reorder_columns = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Date', 'Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours']
df = df[reorder_columns]
df.head()
```

```python
df_mod = df.copy()
```

```python
type(df_mod['Date'][0])
```

```python
df_mod['Date'] = pd.to_datetime(df_mod['Date'],format='%d/%m/%Y')
df_mod['Date']
```

```python
list_month=[]
for i in range(df_mod.shape[0]):
    list_month.append(df_mod['Date'][i].month)
```

```python
df_mod['Month'] = list_month
df_mod
```

```python
def date_to_weekday(date_value):
    return date_value.weekday()
```

```python
df_mod['Day of week'] = df_mod['Date'].apply(date_to_weekday)
```

```python
df_mod.head()
```

## Grouping Education
- Value 0 corresponds to highschool degree
- Value 1 corresponds to postgraduate, PhD or other degree higher than highschool

```python
df_mod['Education'] = df_mod['Education'].map({1:0,2:1,3:1,4:1})
```

```python
df_mod['Education'].value_counts()
```

```python
df_preprocessed = df_mod.copy()
```

```python
df_preprocessed = df_preprocessed.drop(['Date'],axis=1)
```

```python
df_preprocessed.columns.values
```

```python
columns_names_upd = ['Reason_1', 'Reason_2', 'Reason_3', 'Reason_4','Month',
       'Day of week','Transportation Expense', 'Distance to Work', 'Age',
       'Daily Work Load Average', 'Body Mass Index', 'Education',
       'Children', 'Pets', 'Absenteeism Time in Hours' ]
```

```python
df_preprocessed = df_preprocessed[columns_names_upd]
df_preprocessed.shape
df_preprocessed.head()
```

```python
df_preprocessed.to_csv('Absenteeism_preprocessed.csv',index=False)
```
